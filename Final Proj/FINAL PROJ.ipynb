```python
import pandas as pd
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
import os

# Function to check if DataFrame has data
def check_dataframe_content(df):
    return not df.empty

# Function to send an email
def send_email():
    MY_ADDRESS = "dinerosaurrr@outlook.com"
    MY_PASSWORD = "CNaurr123"
    RECIPIENT_ADDRESS = "dinerosaurrr@outlook.com"

    # Email server configuration for Gmail
    HOST_ADDRESS = "smtp.outlook.com"
    HOST_PORT = 587

    # Create the email message
    msg = MIMEMultipart()
    msg['From'] = MY_ADDRESS
    msg['To'] = RECIPIENT_ADDRESS
    msg['Subject'] = "Arbitrage Opportunity Notification"

    body = "You have a Sports Betting Arbitrage Opportunity!"
    msg.attach(MIMEText(body, 'plain'))

    # Save the DataFrame to a CSV file
    filename = "Arbitrage_Opportunities.csv"
    final_df.to_csv(filename, index=False)

    # Attach the CSV file to the email
    attachment = open(filename, "rb")
    part = MIMEBase('application', 'octet-stream')
    part.set_payload(attachment.read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', f"attachment; filename= {filename}")

    msg.attach(part)
    
    try:
        # Establish SMTP connection
        server = smtplib.SMTP(host=HOST_ADDRESS, port=HOST_PORT)
        server.starttls()  # Start TLS encryption
        server.login(MY_ADDRESS, MY_PASSWORD)  # Login with credentials

# Send email
        server.sendmail(MY_ADDRESS, RECIPIENT_ADDRESS, msg.as_string())
        print("Email sent successfully!")
    except Exception as e:
        print(f"Failed to send email. Error: {str(e)}")
    finally:
        server.quit()  # Quit the SMTP session

# Check if DataFrame has data and send email if it does
if check_dataframe_content(final_df):
    send_email()
else:
    print("DataFrame is empty. No email sent.")

```
